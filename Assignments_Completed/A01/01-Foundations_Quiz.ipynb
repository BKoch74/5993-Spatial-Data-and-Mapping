## Q4. File I/O
Write Python code that saves `Hello World` to a file called `hello.txt`, then reads it back and prints the contents.

```python
with open('hello.txt', "w") as file:
    file.write("Hello World")
with open('hello.txt', 'r') as file:
    print(file.read())
```
